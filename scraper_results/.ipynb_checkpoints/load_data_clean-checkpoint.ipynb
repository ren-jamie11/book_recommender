```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_value_counts_in_ranges(series):
    # Define the ranges
    bins = [1, 20, 30, 50, 100, 500, float('inf')]
    labels = ['1-20', '20-30', '30-50', '50-100', '100-500', '500+']
    
    # Cut the series into bins
    binned_series = pd.cut(series, bins=bins, labels=labels, right=False)

    # Get the count of each bin
    binned_counts = binned_series.value_counts().sort_index()

    # Plot the bar chart
    binned_counts.plot(kind='bar', color='skyblue')
    plt.title('Number of Elements in Each Range')
    plt.xlabel('Range')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

all_books = pd.read_parquet("original_books.parquet")
unique_books = all_books.drop_duplicates(subset = [col for col in all_books.columns if col != 'genres'])

user_reviews = pd.read_parquet("user_book_reviews.parquet")
unique_reviews = user_reviews.drop_duplicates(subset = ['title', 'user_id'], keep = 'first')
unique_reviews = unique_reviews.dropna()
```

```python
print("Unique titles", unique_reviews.title.nunique())
print("Unique user_ids", unique_reviews.user_id.nunique())
```

```python
ratings_per_title = unique_reviews.groupby('title')['rating'].count().reset_index(name='num_ratings')
erroneous_titles = ratings_per_title[ratings_per_title.num_ratings > 30].title.values

ratings_per_title = ratings_per_title[~ratings_per_title['title'].isin(erroneous_titles)]
unique_reviews = unique_reviews[~unique_reviews['title'].isin(erroneous_titles)]
unique_books = unique_books[~unique_books['title'].isin(erroneous_titles)]
```

```python
plot_value_counts_in_ranges(unique_reviews.user_id.value_counts())
```

### User item matrix

```python
# Step 1: If duplicates exist for (user_id, title), we must fix them (e.g., by averaging)
user_reviews_grouped = unique_reviews.groupby(['user_id', 'title'])['rating'].mean().reset_index()

# Step 2: Pivot the table
user_item_matrix = user_reviews_grouped.pivot(index='user_id', columns='title', values='rating')

# Step 3: Fill missing values with 0
user_item_matrix = user_item_matrix.fillna(0)
```
